# Data Preprocessing

```python
#defining the features and target for the data
X=df.drop(['Class','id'],axis=1) #Features
y= df['Class']            #Target
```

Our variable are defined and now we can procced with model development but first we need to scale the data in X as the amount and other columns V1,v2,v3 etc are not upto scale to get compared.

```python
#scaling features to be compared
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
#splitting data into training and testing data
X_train,X_test,y_train,y_test = train_test_split(X_scaled,y,test_size=0.2,random_state=7)
```

# Models

## Decision Tree

```python
#fitting the model
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier(max_depth = 3, random_state=42)
dtc.fit(X_train,y_train)
```

```python
#getting the prediction
y_pred = dtc.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

#accuracy scorw
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")

print(f"classificiation Report:\n{classification_report(y_test, y_pred, target_names=['Non-Fraud', 'Fraud'])}")
```

```python
# Generate the confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
conf_matrix
```

The model performed well on this datset and have a good accuracy score of 0.9475 and fraud recall  as 0.93 with not many mistakes

```python
#plotting the the decesion to know how decision was made

import graphviz
from sklearn.tree import export_graphviz
from sklearn import tree
# Export the tree to a dot file
dot_data = tree.export_graphviz(dtc,
                                out_file=None,
                                feature_names=X.columns,
                                class_names=['Non-Fraud', 'Fraud'],
                                filled=True,
                                rounded=True,
                                special_characters=True)

# Render the dot file into a graph
graph = graphviz.Source(dot_data)

# Display the graph
graph.render("decision_tree")  # Saves the tree as 'decision_tree.pdf'
graph
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=50, random_state=42, max_depth=5)
rfc.fit(X_train, y_train)

# Make predictions on the test set
y_pred2 = rfc.predict(X_test)

# Calculate the accuracy
accuracy = accuracy_score(y_test, y_pred2)
print(f"Accuracy: {accuracy:.4f}")

# Print classification report
print(f"Classification Report:\n{classification_report(y_test, y_pred2, target_names=['Non-Fraud', 'Fraud'])}")
```

```python
# Generate the confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred2)
conf_matrix
```

it is slight better than decision tree with accuracy of 0.9533 and with fraud recall of 0.91 but took a long time to execute haveing 50 trees and 5 depth.

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
# Create and train the Logistic Regression model
log_reg = LogisticRegression(max_iter=1000, random_state=42)
log_reg.fit(X_train, y_train)

# Make predictions on the test set
y_pred3 = log_reg.predict(X_test)

# Calculate the accuracy
accuracy = accuracy_score(y_test, y_pred3)
print(f"Accuracy: {accuracy:.4f}")

# Print classification report
print(f"Classification Report:\n{classification_report(y_test, y_pred3, target_names=['Non-Fraud', 'Fraud'])}")
```

```python
# Generate the confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred3)
conf_matrix
```

Logisistic regression performed better overall than previous algorithm with an accuracy score of 0.9645 and executed in much less time than other also with great fraud recall of 0.95.

## Gaussian Naive Bayes Model

```python
from sklearn.naive_bayes import GaussianNB

# Create and train the Gaussian Naive Bayes model
gnb = GaussianNB()
gnb.fit(X_train, y_train)

# Make predictions on the test set
y_pred4 = gnb.predict(X_test)
```

so this model the acurracy score is worse than the other here and can be a bad choice compared to others and its fraud recall is also bad here.

## Nearest Neighbor

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

y_pred6 = knn.predict(X_test)

# Calculate the accuracy
accuracy = accuracy_score(y_test, y_pred6)
print(f"Accuracy: {accuracy:.4f}")

# Print classification report
print(f"Classification Report:\n{classification_report(y_test, y_pred6, target_names=['Non-Fraud', 'Fraud'])}")
```

```python
# Generate the confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred6)
conf_matrix
```

By Nearest Neighbor we got the best result now with an outstanding accuracy of 0.9982 accuracy and with perfect fraud recall 1.0 But took super long time to execute.

# Conclusion

We compared Accuracy Score for Every model and till now decide on the basis of Accuracy Score only. So according to Accuracy if you want to classify more accurately then KNN model will be better. But practically;we  have  to minimize false negative that the credit card fraud classified as not fraud because if a fraud classified as non fraud can cause harm so we have you choose the model with least false negative which can be done by recall or by corr matrix So the model with the highesr Fraud Recall is given by Logistic Regression with value = 1.0. But In practicality we also need the model to work properly and customer to wait little but KNN took much longer time than any other Model which can cause the system to classify a transaction Fraud or not fraud take long time which can be impractical considering no. of transaction in real time. So we can choose Logistic Regression as it is much faster and have a pretty good recall score of 0.95. So, we can choose Logistic Regression to be deployed Model.
